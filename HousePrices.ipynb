```python
train_n=train.shape[0]
test_n=test.shape[0]
y_train=train.SalePrice.values
y_test=train['SalePrice']
all_data=pd.concat((train,test), sort=False).reset_index(drop=True)
all_data.drop(['SalePrice'], axis=1, inplace=True)
all_data.shape
```

```python
#getting missing data
all_data_na_values=all_data.isnull().sum()
all_data_na_values=all_data_na_values.drop(all_data_na_values[all_data_na_values==0].index).sort_values(ascending=False)[:30]
all_data_na_values
```

```python
plt.subplots(figsize=(15,12))
plt.xticks(rotation='90')
sns.barplot(x=all_data_na_values.index,y=all_data_na_values)
plt.xlabel('Features')
plt.ylabel('Missing Values')
plt.title('Missing data by Features')
```

```python
#correlation map to see how features are correlated with SalePrice
corrmat= train.corr()
plt.subplots(figsize= (25,15))
sns.heatmap(corrmat, vmax= 0.9, square= True, annot= True, fmt= ".2f")
```

```python
#fill missing values
all_data['PoolQC']= all_data['PoolQC'].fillna("None")
all_data['MiscFeature']= all_data['MiscFeature'].fillna("None")
all_data["Alley"]= all_data["Alley"].fillna("None")
all_data["Fence"]= all_data["Fence"].fillna("None")
all_data["FireplaceQu"]= all_data["FireplaceQu"].fillna("None")
all_data["LotFrontage"]= all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))

for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    all_data[col]= all_data[col].fillna('None')
    
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    all_data[col]= all_data[col].fillna(0)
    
for col in ('BsmtCond', 'BsmtExposure', 'BsmtQual', 'BsmtFinType2', 'BsmtFinType1'):
    all_data[col]= all_data[col].fillna('None')
    
for col in ('BsmtHalfBath', 'BsmtFullBath', 'TotalBsmtSF', 'BsmtUnfSF', 'BsmtFinSF1', 'BsmtFinSF2'):
    all_data[col]= all_data.fillna(0)
    
all_data["MasVnrType"]= all_data["MasVnrType"].fillna("None")
all_data["MasVnrArea"]= all_data["MasVnrArea"].fillna(0)
all_data["MSZoning"]= all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0])
all_data= all_data.drop(['Utilities'], axis= 1)
all_data["Functional"]= all_data["Functional"]
all_data["Electrical"]= all_data["Electrical"].fillna(all_data['Electrical'].mode()[0])
all_data["KitchenQual"]= all_data["KitchenQual"].fillna(all_data['KitchenQual'].mode()[0])
all_data["Exterior1st"]= all_data["Exterior1st"].fillna(all_data["Exterior1st"].mode()[0])
all_data["Exterior2nd"]= all_data["Exterior2nd"].fillna(all_data["Exterior2nd"].mode()[0])
all_data["SaleType"]= all_data["SaleType"].fillna(all_data["SaleType"].mode()[0])
```

```python
#MSSubClass=The building class
all_data['MSSubClass']= all_data['MSSubClass'].apply(str)

#changing OverallCond into a categorical variable
all_data['OverallCond']= all_data['OverallCond'].astype(str)

#Year and month sold are transformed into categorical features
all_data['YrSold']= all_data['YrSold'].astype(str)
all_data['MoSold']= all_data['MoSold'].astype(str)
```

```python
cols=('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 'ExterQual', 'ExterCond', 'HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope', 'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'YrSold', 'MoSold')

#process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl= LabelEncoder()
    lbl.fit(list(all_data[c].values))
    all_data[c]= lbl.transform(list(all_data[c].values))

all_data.shape
```

```python
#adding total sqfoot feature
all_data['TotalSF']= all_data['TotalBsmtSF'] + all_data['1stFlrSF'] + all_data['2ndFlrSF']
```

```python
num= all_data.dtypes[all_data.dtypes!= 'object'].index

#skew all numerical features
skew_feat= all_data[num].apply(lambda x: skew(x.dropna())).sort_values(ascending= False)

sk= pd.DataFrame({'Skewness' :skew_feat})
sk.head(10)
```

```python
#getting dummy categorical features
#should/need to define categorical columns list
all_data= pd.get_dummies(all_data)
print(all_data.shape)
```

```python
train_new= all_data[:train_n]
train_new.shape
```

```python
test_new= all_data[train_n:]
test_new.shape
```

```python
regr= xgb.XGBRegressor(colsample_bytree= 0.2, gamma= 0.0, learning_rate= 0.1, max_depth= 4, min_child_weight= 1.5, n_estimators= 7200, reg_alpha= 0.9, reg_lambda= 0.6, subsample= 0.2, seed= 42, silent= 1)
regr.fit(train_new, y_train)
y_pred= regr.predict(train_new)
y_test= train['SalePrice']

score_xgb= np.sqrt(mean_squared_error(y_test, y_pred))

y_pred_xgb= regr.predict(test_new)
y_pred_xgb= np.exp(y_pred_xgb)

score_xgb
```

```python
model_lasso= Lasso(alpha= 0.0005, normalize= True, max_iter= 1e5)
model_lasso.fit(train_new, y_train)
y_pred_lasso= model_lasso.predict(train_new)
score_lasso= np.sqrt(mean_squared_error(y_train, y_pred_lasso))


y_pred_lasso_pred= model_lasso.predict(test_new)
y_pred_lasso_pred= np.exp(y_pred_lasso_pred)

score_lasso
```

```python
model_rd= Ridge(alpha= 4.84)
model_rd.fit(train_new, y_train)
y_pred_rd= model_rd.predict(train_new)
score_rd= np.sqrt(mean_squared_error(y_train, y_pred_rd))

y_pred_rd_pred= model_rd.predict(test_new)
y_pred_rd_pred= np.exp(y_pred_rd_pred)

score_rd
```

```python
model_enet= ElasticNet(alpha= 0.0005, l1_ratio= 0.9, random_state= 3)
model_enet.fit(train_new, y_train)
y_pred_enet= model_enet.predict(train_new)
score_enet= np.sqrt(mean_squared_error(y_train, y_pred_enet))

y_pred_enet_pred= model_enet.predict(test_new)
y_pred_enet_pred= np.exp(y_pred_enet_pred)

score_enet
```

```python
model_rf= RandomForestRegressor(n_estimators= 12, max_depth= 3, n_jobs= -1)
model_rf.fit(train_new, y_train)
y_pred_rf= model_rf.predict(train_new)
score_rf= np.sqrt(mean_squared_error(y_train, y_pred_rf))

y_pred_rf_pred= model_rf.predict(test_new)
y_pred_rf_pred= np.exp(y_pred_rf_pred)

score_rf
```

```python
model_gb= GradientBoostingRegressor(n_estimators= 40, max_depth= 2)
model_gb.fit(train_new, y_train)
y_pred_gb= model_gb.predict(train_new)
score_gb= np.sqrt(mean_squared_error(y_train, y_pred_gb))

y_pred_gb_pred= model_gb.predict(test_new)
y_pred_gb_pred= np.exp(y_pred_gb_pred)

score_gb
```

```python
dt= DecisionTreeRegressor(max_depth=4)
model_ad= AdaBoostRegressor(dt, learning_rate= 0.1, n_estimators= 300, random_state= None)
model_ad.fit(train_new, y_train)
y_pred_ad= model_ad.predict(train_new)
score_ad= np.sqrt(mean_squared_error(y_train, y_pred_ad))

y_pred_ad_pred= model_ad.predict(test_new)
y_pred_ad_pred= np.exp(y_pred_ad_pred)

print(score_ad)
```

```python
avg= (y_pred_xgb + y_pred_lasso_pred + y_pred_rd_pred + y_pred_enet_pred + y_pred_rf_pred + y_pred_gb_pred + y_pred_ad_pred) / 7
```

```python
avgPred= pd.DataFrame()
avgPred['Id']= test_ID
avgPred['SalePrice']= avg
avgPred.to_csv('house_price_predictions.csv', index= False)
```
